```python
import pandas as pd
```

```python
# Открываем файл для прочтения
df = pd.read_csv('Desktop/horse_data.csv', sep = ',', header=0, na_values='?')
```

```python
# Смотрим что внутри
df.info()
```

```python
# Столбцы не имеют нормальных названий, необходимо их заменить 
df.head()
```

```python
# Меняем названия на те, что указаны в документации
df.columns = ['surgery', 'age', 'Hospital_Number', 'rectal_temperature', 'pulse', 'respiratory_rate', 'temperature_of_extremities', 'peripheral_pulse', 'mucous_membranes', 'capillary_refill_time', 'pain', 'peristalsis', 'abdominal_distension', 'nasogastric_tube', 'nasogastric_reflux', 'nasogastric_reflux_PH', 'rectal_examination_feces', 'abdomen', 'packed_cell_volume', 'total_protein', 'abdominocentesis_appearance', 'abdomcentesis_total_protein', 'outcome', 'surgical_lesion', '25type_of_lesion', '26type_of_lesion', '27type_of_lesion', 'cp_data']
df.head()
```

```python
# Оставляем только те столбцы, которые нас интересуют
df_filter = df[['surgery', 'age', 'rectal_temperature', 'pulse', 'respiratory_rate', 'temperature_of_extremities', 'pain', 'outcome']]
```

```python
# Проверяем
df_filter.head()
```

```python
# Уникальные значения столбцов
print(df_filter['surgery'].unique())
print(df_filter['age'].unique())
print(df_filter['rectal_temperature'].unique())
print(df_filter['pulse'].unique())
print(df_filter['respiratory_rate'].unique())
print(df_filter['temperature_of_extremities'].unique())
print(df_filter['pain'].unique())
print(df_filter['outcome'].unique())
```

```python
# Стоит обратить внимание на столбец "age". Судя по документации, условное обозначение возраста должно быть 1 и 2.
# Значит 9 - это ошибика. Нужно заменить значения на корректные.
```

```python
df_filter['age'] = df_filter.age.apply(lambda x: 2 if x == 9 else x)
df_filter
```

```python
# Исправляем типы данных
df_filter['surgery'] = df_filter['surgery'].astype('float').astype('Int64')
df_filter['pain'] = df_filter['pain'].astype('float').astype('Int64')
df_filter['outcome'] = df_filter['outcome'].astype('float').astype('Int64')
df_filter
```

```python
# Базовые статистики
df_filter.describe()
```

```python
# Моды
print('surgery', df_filter['surgery'].mode()[0])
print('age', df_filter['age'].mode()[0])
print('rectal_temperature', df_filter['rectal_temperature'].mode()[0])
print('pulse', df_filter['pulse'].mode()[0])
print('respiratory_rate', df_filter['respiratory_rate'].mode()[0])
print('temperature_of_extremities', df_filter['temperature_of_extremities'].mode()[0])
print('pain', df_filter['pain'].mode()[0])
print('outcome', df_filter['outcome'].mode()[0])
```

```python
# Дисперсии
print('surgery', df_filter['surgery'].var())
print('age', df_filter['age'].var())
print('rectal_temperature', df_filter['rectal_temperature'].var())
print('pulse', df_filter['pulse'].var())
print('respiratory_rate', df_filter['respiratory_rate'].var())
print('temperature_of_extremities', df_filter['temperature_of_extremities'].var())
print('pain', df_filter['pain'].var())
print('outcome', df_filter['outcome'].var())
```

```python
# Медианы
print('surgery_median', df_filter['surgery'].median())
print('age_median', df_filter['age'].median())
print('rectal_temperature_median', df_filter['rectal_temperature'].median())
print('pulse_median', df_filter['pulse'].median())
print('respiratory_rate_median', df_filter['respiratory_rate'].median())
print('temperature_of_extremities_median', df_filter['temperature_of_extremities'].median())
print('pain_median', df_filter['pain'].median())
print('outcome_median', df_filter['outcome'].median())
```

```python
# Межквартильные размахи

Q1_rectal_temperature = df_filter['rectal_temperature'].quantile(0.25)
Q3_rectal_temperature = df_filter['rectal_temperature'].quantile(0.75)
IQR_rectal_temperature = Q3_rectal_temperature - Q1_rectal_temperature
print('rectal_temperatur', IQR_rectal_temperature)

Q1_pulse = df_filter['pulse'].quantile(0.25)
Q3_pulse = df_filter['pulse'].quantile(0.75)
IQR_pulse = Q3_pulse - Q1_pulse
print('pulse', IQR_pulse)

Q1_respiratory_rate = df_filter['respiratory_rate'].quantile(0.25)
Q3_respiratory_rate = df_filter['respiratory_rate'].quantile(0.75)
IQR_respiratory_rate = Q3_respiratory_rate - Q1_respiratory_rate
print('respiratory_rate', IQR_respiratory_rate)
```

```python
# Выбросы
# Я искала выбросы только в трех столбцах: температура, дыхание и пульс, 
# так как только в них есть значения, а не категории.

lower_bound_rectal_temperature = Q1_rectal_temperature - (1.5 * IQR_rectal_temperature) 
upper_bound_rectal_temperature = Q3_rectal_temperature + (1.5 * IQR_rectal_temperature)
remove_outliers_rectal_temperature = df_filter[df_filter['rectal_temperature']\
                                    .between(lower_bound_rectal_temperature, upper_bound_rectal_temperature, inclusive = True)]\
                                    .sort_values('outcome')
outliers_rectal_temperature = pd.concat([df_filter, remove_outliers_rectal_temperature]).drop_duplicates(keep = False)
outliers_rectal_temperature.head()
```

```python
outliers_rectal_temperature['rectal_temperature'].value_counts()
```

```python
lower_bound_pulse = Q1_pulse - (1.5 * IQR_pulse) 
upper_bound_pulse = Q3_pulse + (1.5 * IQR_pulse)
remove_outliers_pulse = df_filter[df_filter['pulse']\
                        .between(lower_bound_pulse, upper_bound_pulse, inclusive = True)].sort_values('outcome')
outliers_pulse = pd.concat([df_filter, remove_outliers_pulse]).drop_duplicates(keep = False)
outliers_pulse.head()
```

```python
outliers_pulse['pulse'].value_counts()
```

```python
lower_bound_respiratory_rate = Q1_respiratory_rate - (1.5 * IQR_respiratory_rate) 
upper_bound_respiratory_rate = Q3_respiratory_rate + (1.5 * IQR_respiratory_rate)
remove_outliers_respiratory_rate = df_filter[df_filter['respiratory_rate']\
                                   .between(lower_bound_respiratory_rate, upper_bound_respiratory_rate, inclusive = True)]\
                                   .sort_values('outcome')
outliers_respiratory_rate = pd.concat([df_filter, remove_outliers_respiratory_rate]).drop_duplicates(keep = False)
outliers_respiratory_rate.head()
```

```python
outliers_respiratory_rate['respiratory_rate'].value_counts()
```

```python
# Работа с пропусками
df_filter_v1 = df_filter.copy()
df_filter_v1
```

```python
df_filter_v1.info()
```

```python
# Нужно оценить долю пропусков по столбцам.
(df_filter_v1.isna().mean() * 100).round(2)
# Столбец age не имеет пропусков и все данные соответствуют документации.
# В столбцах surgery и outcome отсутствует одинаковое количество пропусков
# rectal_temperature - самый большой процент пропусков, по-этому игнорировать их нельзя, а собрать данные, очевидно, мы не можем. 
# Нужно будет заменить средними значениями. Я брала моду, как меру центральной тенденции. 
# В слобцах:
#pulse 
# respiratory_rate 
# temperature_of_extremities 
# pain присутствуют выбросы, стоит заменить пропуски на значение медианы.
```

```python
df_filter_v1['outcome'].fillna(df_filter_v1['outcome'].mode()[0], inplace=True)
print(df_filter_v1['outcome'].isna().sum())
print(df_filter_v1['outcome'].unique())
```

```python
df_filter_v1['pain'].fillna(df_filter_v1['pain'].mode()[0], inplace=True)
print(df_filter_v1['pain'].isna().sum())
print(df_filter_v1['pain'].unique())
```

```python
df_filter_v1['rectal_temperature'].fillna(df_filter_v1['rectal_temperature'].median(), inplace=True)
df_filter_v1['rectal_temperature']
```

```python
df_filter_v1['pulse'].fillna(df_filter_v1['pulse'].median(), inplace=True)
df_filter_v1['pulse']
```

```python
df_filter_v1['respiratory_rate'].fillna(df_filter_v1['respiratory_rate'].median(), inplace=True)
df_filter_v1['respiratory_rate']
```

```python
df_filter_v1['temperature_of_extremities'].fillna(df_filter_v1['temperature_of_extremities'].median(), inplace=True)
df_filter_v1['temperature_of_extremities']
```
