```python
df_heart['age'].plot(kind = 'hist', title = 'Риск ССЗ по возрастным группам')
```

```python
# Люди возрастом около 60и лет болеют ССЗ чаще, чем остальные.
# Можно выдвинуть гипотезу о том, как возрастные изменения влияют на вероятность заболеваний сердца. 
```

```python
age_groups = df_heart[['age', 'DEATH_EVENT']].groupby('age').sum().sort_values('DEATH_EVENT', ascending = False)
age_groups.head()
```

```python
#Смертельные случаи в возрастных категориях
age_groups.style.bar(align = 'mid')
```

```python
# Поверхностный взгляд на данные дает возможность предположить, что в возрастной группе 60+- 
# риск смертельного исхода при заболевании ССЗ очень высок. 
# Можно выдвинуть гипотезу о зависимости смертельных случаев от возраста пациента. 
```
